```python
from cvid_data import cvid_2d as cvid
import numpy as np 
import os
```

```python
Ku10 = os.listdir("/mnt/raid-cita/dongwooc/20210519_limlam_Ku10/")
Ka21 = os.listdir("/mnt/raid-cita/dongwooc/20210519_limlam_Ka21/")
Ka10 = os.listdir("/mnt/raid-cita/dongwooc/20210519_limlam_Ka10/")

intersect = set(Ka21).intersection(set(Ka10)) #common files
 
Ku10_ = []
for npz in Ku10:
    if npz in intersect:
        Ku10_.append(npz)        
Ka21_ = []
for npz in Ka21:
    if npz in intersect:
        Ka21_.append(npz)         
Ka10_ = []
for npz in Ka10:
    if npz in intersect:
        Ka10_.append(npz)
        
#seperate smooth and unsmooth        
def make_lists(lst):
    ls = []
    unsmooth_ls = []
    for npz in lst:
        if 'unsmooth' in npz:
            unsmooth_ls.append(npz)
        else:
            ls.append(npz)
    return ls, unsmooth_ls
Ku10_s, Ku10_us = make_lists(Ku10_)
Ka21_s, Ka21_us = make_lists(Ka21_)
Ka10_s, Ka10_us = make_lists(Ka10_)

#triplets
s = []
us = []
for npz in Ku10_s:
    s.append(npz) 
    s.append(Ka21_s[Ka21_s.index(npz)]) 
    s.append(Ka10_s[Ka10_s.index(npz)])
s = np.reshape(np.array(s), (len(Ku10_s),3)) 

for npz in Ku10_us:
    us.append(npz)
    us.append(Ka21_us[Ka21_us.index(npz)]) 
    us.append(Ka10_us[Ka10_us.index(npz)])
us = np.reshape(np.array(us), (len(Ku10_us),3))
```

```python
smooth = s[:80]
T = []
for triplet in smooth:
    os.chdir("/mnt/raid-cita/dongwooc/20210519_limlam_Ku10/") 
    S1 = np.load(triplet[0])['map_cube']
    os.chdir("/mnt/raid-cita/dongwooc/20210519_limlam_Ka21/") 
    S2 = np.load(triplet[1])['map_cube']
    Tmax = max(S1.flatten().max(),S2.flatten().max()) + 5
    T.append(Tmax)
t = max(T)
```

```python
class get_cvid_data:
    def __init__(self,S1,S2,S3):
       self.S1 = 0*S1
       self.S2 = 0*S2
       self.S3 = S3
       global Nbin 
       Nbin = 51
       global N_resample 
       N_resample = 50
       global Tmax
       Tmax = 30*t
       global m
       m = cvid(self.S1.flatten(),self.S2.flatten(),Nbin1=Nbin,Nbin2=None,
                N_resample=N_resample,Tmax1=Tmax,Tmax2=Tmax,fTmax1=None,fTmax2=None)
       global mm
       mm = cvid(self.S1.flatten() - np.mean(self.S1.flatten()),
                  self.S2.flatten() - np.mean(self.S2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)
       global msn
       N1 = np.random.normal(loc=0., scale=3.03, size=(30,30,512)) 
       N2 = np.random.normal(loc=0., scale=4.93, size=(30,30,512)) 
       M1 = self.S1+N1
       M2 = self.S2+N2
       msn = cvid(M1.flatten() - np.mean(M1.flatten()),
                  M2.flatten() - np.mean(M2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)
       global mco 
       M1 = self.S1+self.S3+N1
       mco = cvid(M1.flatten() - np.mean(M1.flatten()),
                  M2.flatten() - np.mean(M2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)   
        
       global CO10_05
       M1 = 0.5*self.S1+self.S3+N1
       CO10_05 = cvid(M1.flatten() - np.mean(M1.flatten()),
                  M2.flatten() - np.mean(M2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)
       global CO10_2
       M1 = 2*self.S1+self.S3+N1
       CO10_2 = cvid(M1.flatten() - np.mean(M1.flatten()),
                  M2.flatten() - np.mean(M2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)
       global CO10_3
       M1 = 3*self.S1+self.S3+N1
       CO10_3 = cvid(M1.flatten() - np.mean(M1.flatten()),
                  M2.flatten() - np.mean(M2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)
    
       M1 = self.S1+self.S3+N1
       global CO21_05
       M2 = 0.5*self.S2+N2
       CO21_05 = cvid(M1.flatten() - np.mean(M1.flatten()),
                  M2.flatten() - np.mean(M2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)
       global CO21_2
       M2 = 2*self.S2+N2
       CO21_2 = cvid(M1.flatten() - np.mean(M1.flatten()),
                  M2.flatten() - np.mean(M2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)
       global CO21_3
       M2 = 3*self.S2+N2
       CO21_3 = cvid(M1.flatten() - np.mean(M1.flatten()),
                  M2.flatten() - np.mean(M2.flatten()),
                  Nbin1=Nbin,Nbin2=None,N_resample=N_resample,Tmax1=Tmax,
                  Tmax2=Tmax,fTmax1=None,fTmax2=None)
    def mRr(self):         
        return m.R.real #(25,50)
    def mRi(self):
        return m.R.imag
    def mRrerr(self):
        return m.Rerr[0:1250].reshape(50,25)
    def mRierr(self):
        return m.Rerr[1250:].reshape(50,25)
    def mfT1(self):
        return m.fT1
    def mfT2(self):
        return m.fT2
    def mmRr(self):
        return mm.R.real
    def mmRi(self):
        return mm.R.imag
    def mmRrerr(self):
        return mm.Rerr[0:1250].reshape(50,25)
    def mmRierr(self):
        return mm.Rerr[1250:].reshape(50,25)
    def msnRr(self):
        return msn.R.real
    def msnRi(self):
        return msn.R.imag
    def msnRrerr(self):
        return msn.Rerr[0:1250].reshape(50,25)
    def msnRierr(self):
        return msn.Rerr[1250:].reshape(50,25)
    def mcoRr(self):
        return mco.R.real
    def mcoRi(self):
        return mco.R.imag
    def mcoRrerr(self):
        return mco.Rerr[0:1250].reshape(50,25)
    def mcoRierr(self):
        return mco.Rerr[1250:].reshape(50,25)
    #CO10_05
    def CO10_05Rr(self):
        return CO10_05.R.real
    def CO10_05Ri(self):
        return CO10_05.R.imag
    def CO10_05Rrerr(self):
        return CO10_05.Rerr[0:1250].reshape(50,25)
    def CO10_05Rierr(self):
        return CO10_05.Rerr[1250:].reshape(50,25)
    #CO10_2
    def CO10_2Rr(self):
        return CO10_2.R.real
    def CO10_2Ri(self):
        return CO10_2.R.imag
    def CO10_2Rrerr(self):
        return CO10_2.Rerr[0:1250].reshape(50,25)
    def CO10_2Rierr(self):
        return CO10_2.Rerr[1250:].reshape(50,25)
    #CO10_3
    def CO10_3Rr(self):
        return CO10_3.R.real
    def CO10_3Ri(self):
        return CO10_3.R.imag
    def CO10_3Rrerr(self):
        return CO10_3.Rerr[0:1250].reshape(50,25)
    def CO10_3Rierr(self):
        return CO10_3.Rerr[1250:].reshape(50,25)
    #CO21_05
    def CO21_05Rr(self):
        return CO21_05.R.real
    def CO21_05Ri(self):
        return CO21_05.R.imag
    def CO21_05Rrerr(self):
        return CO21_05.Rerr[0:1250].reshape(50,25)
    def CO21_05Rierr(self):
        return CO21_05.Rerr[1250:].reshape(50,25)
    #CO21_2
    def CO21_2Rr(self):
        return CO21_2.R.real
    def CO21_2Ri(self):
        return CO21_2.R.imag
    def CO21_2Rrerr(self):
        return CO21_2.Rerr[0:1250].reshape(50,25)
    def CO21_2Rierr(self):
        return CO21_2.Rerr[1250:].reshape(50,25)
    #CO21_3
    def CO21_3Rr(self):
        return CO21_3.R.real
    def CO21_3Ri(self):
        return CO21_3.R.imag
    def CO21_3Rrerr(self):
        return CO21_3.Rerr[0:1250].reshape(50,25)
    def CO21_3Rierr(self):
        return CO21_3.Rerr[1250:].reshape(50,25)
```

```python
for triplet in smooth:
    os.chdir("/mnt/raid-cita/dongwooc/20210519_limlam_Ku10/")
    Ku10 = np.load(triplet[0])
    os.chdir("/mnt/raid-cita/dongwooc/20210519_limlam_Ka21/")
    Ka21 = np.load(triplet[1])
    os.chdir("/mnt/raid-cita/dongwooc/20210519_limlam_Ka10/")
    Ka10 = np.load(triplet[2])

    a = get_cvid_data(Ku10['map_cube'],
                      Ka21['map_cube'],
                      Ka10['map_cube'])

    os.chdir("/cita/d/homes/home-2/bangari/repo")
    np.savez(triplet[0][:6]+'_cvid_data0', 
             mRr=a.mRr(),mRi=a.mRi(),mRrerr=a.mRrerr(),mRierr=a.mRierr(),mfT1=a.mfT1(),mfT2=a.mfT2(),
             mmRr=a.mmRr(),mmRi=a.mmRi(),mmRrerr=a.mmRrerr(),mmRierr=a.mmRierr(),
             msnRr=a.msnRr(),msnRi=a.msnRi(),msnRrerr=a.msnRrerr(),msnRierr=a.msnRierr(),
             mcoRr=a.mcoRr(),mcoRi=a.mcoRi(),mcoRrerr=a.mcoRrerr(),mcoRierr=a.mcoRierr()
            )
```
